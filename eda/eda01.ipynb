# EDA - Exploring Tables and Counts

```python
import pandas as pd
import sqlite3
```

### Creating Connection

```python
conn = sqlite3.connect("vic_crash_data.db")
```

### Accidents table
Contains meta information about accident date, time, week, vehicles involved, location, people injured, and target variable severity

```python
accidents = pd.read_sql("select * from accidents limit 5", conn)
accidents
```

### Vehicle Table
Contains all vehicle information involved in the accident - there can be multiple vehicles involved in an accident, so an accident_no can have multiple vehicles involved

```python
vehicles = pd.read_sql("Select * from vehicle_info limit 5", conn)
vehicles
```

```python
count_of_vehicles = pd.read_sql("""
select count(distinct v.vehicle_id) count_of_vehicles, v.accident_no from vehicle_info as v
group by v.accident_no
limit 5
""", conn)
count_of_vehicles
```

### Person Table
Contains information about people involved in the accident, their age group, injuries and whether taken to hospital

```python
person = pd.read_sql("Select * from person_info limit 5", conn)
person
```

### Table Accident_Node
Contains geographical location of accident, a node_id can have multiple accidents

```python
nodes = pd.read_sql("Select * from accident_node limit 5", conn)
nodes
```

```python
count_of_accidents = pd.read_sql("""
select count(distinct a.accident_no) count_of_accidents, a.node_id from accident_node a
group by a.node_id
order by count_of_accidents desc
limit 5
""", conn)
count_of_accidents
```
